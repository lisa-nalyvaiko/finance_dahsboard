```python
key = '20a5ec0315fca4f5798172d3f11be2fdf3e36b37' #BambooHR api key
url_leaves = 'https://api.bamboohr.com/api/gateway.php/beetroot/v1/time_off/requests/' 
headers = {"Accept": "application/json"}
start_date = "2021-01-01" #stating which timeframe we need the data for
end_date = "2021-01-31"
querystring = {"start":start_date,"end":end_date}
```

```python
response_leaves = requests.get(url_leaves,headers=headers,params=querystring, auth=(key, 'x'))
```

```python
response_leaves.text
```

```python
#cleaning the data
leaves = pd.read_json(response_leaves.text)
leaves
```

```python
leaves[['start','end','created']] = leaves[['start','end','created']].apply(pd.to_datetime)
```

```python
startdate_datetime = datetime.strptime(start_date, '%Y-%m-%d')
enddate_datetime = datetime.strptime(end_date, '%Y-%m-%d')
```

```python
#changing dates in leaves from strings to datetime
upd_dates = []
for i in leaves.dates:
    upd_date = {(datetime.strptime(k, '%Y-%m-%d')):float(v) for (k,v) in i.items()}
    upd_dates.append(upd_date)
```

```python
#removing PADs that don't fit in our timeframe
final_leaves = []
for i in upd_dates:
    final_leave = {k:v for (k,v) in i.items() if k>=startdate_datetime and k<=enddate_datetime}
    final_leaves.append(final_leave)
```

```python
#geting sum of leaves
leaves.amount = [sum(i.values()) for i in final_leaves]
```

```python
leaves
```

```python
#removing requests that didn't get approved
leaves.status = leaves.status.apply(pd.Series)['status']
```

```python
leaves.drop(leaves[leaves['status'] == 'superceded'].index,inplace=True)
```

```python
leaves = leaves.drop(columns=['id','status','created','notes','actions'])
```

```python
leaves_summary = leaves[['name','amount']].groupby('name',as_index=False).agg({'amount':'sum'})
```

```python
leaves_summary
```

```python
#getting current salaries from BambooHR
url_salaries = "https://api.bamboohr.com/api/gateway.php/beetroot/v1/reports/276"
querystring1 = {"format":"CSV"}
headers = {"Accept": "application/json"}
response_salaries = requests.get(url_salaries, headers=headers, params=querystring,auth=(key, 'x'))
```

```python
report = json.loads(response_salaries.content)
salaries = pd.DataFrame(report['employees'])
salaries
```

```python
salaries = salaries.rename(columns={'fullName2':'full_name','payRateEffectiveDate':'last_salary_change','payRate':'salary_$'})
```

```python
salaries['name'] = salaries.firstName + ' ' + salaries.lastName
```

```python
#dropping unnecessary columns
salaries = salaries.drop(columns=['full_name','firstName','lastName','id'])
```

```python
#removing $ sign from salaries
salary = []
s = salaries['salary_$'].str.split('.')
for i in s:
    salary.append(i[0])
salaries['salary_$'] = salary
```

```python
#fixing roles names
salaries.jobTitle = salaries.jobTitle.str.replace('WordPress developer\Team Lead','Tech Lead',regex=False)
salaries.jobTitle = salaries.jobTitle.str.replace('WordPress developer\Tech Lead','Tech Lead',regex=False)
```

```python
salaries
```

```python
#joining salary data with PADs data
team_stats = salaries.merge(leaves_summary,how='left',on='name')
```

```python
team_stats
```

```python
team_stats.amount = team_stats.amount.fillna(0)
```

```python
#setting the last month
last_month = datetime.now() - relativedelta(months=1)*2
#checking if we got it right
text_last_month = format(last_month, '%B %Y')
text_last_month
```

```python
#counting the total working hours amount for last month and working hours per person
working_days = np.busday_count(start_date, end_date)
total_working_h = working_days * 8
team_stats['working_h'] = total_working_h - team_stats['amount'] * 8
working_h_summary = f"Total working hours in {text_last_month} = {total_working_h}h"
print(working_h_summary)
team_stats = team_stats.rename(columns={'amount':'leave_days'})
```

```python
#sorting the data by location and department
team_stats = team_stats.sort_values(by=['location','department','name'])
```
